 #### **Write all numeric veriable and make them numeric**
1. `Size`
2. ``Install``
3. `Price`   


Check that how many MBs are in one KB? And how to calculate? 
  1.  ## Size
  . Covert KBs into MBs \
  . Then remove M from all numbers \
  . Handle `varies eith device`


```python
df['Size'].isnull().sum()
```

```python
df['Size'].unique()
```

```python
df['Size'] = df['Size'].astype(str).str.replace('k','M').astype(float) * 1024 / 1024
```

```python
df['Size'].value_counts()
```

2. ## Install 

```python
df['Installs'].isnull().sum()
```

```python
df['Installs'].value_counts()
```

3. ## Price 
    1. Remove $ sign from Price

```python
df['Price'].isnull().sum()
```

```python
df['Price'].value_counts()
```

```python
# convert price object to numeric
df['Price'] = pd.to_numeric(df['Price'].replace('$','') ,errors='coerce' ) 
```

```python
#convert installs object to numeric
# df['Installs'] = pd.to_numeric(df['Installs'].str.replace('+', '').str.replace(',', ''), errors='coerce') 
df['Installs'] = pd.to_numeric(df['Installs'].str.replace('+','').str.replace(',','') , errors='coerce')
```

```python
df['Size']  = pd.to_numeric(df['Size'].str.replace('M','') , errors='coerce')
```
