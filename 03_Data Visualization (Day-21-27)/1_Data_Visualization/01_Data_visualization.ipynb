```python
# Import libraries
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import titanic dataset from seaborn library
df = sns.load_dataset('titanic')
# Print the first 5 rows of the dataset
print(df.head())
```

```python
df.head()
```

```python
# lets create a scatter plot using df 
df.plot(kind="scatter" , x = 'age' , y = 'fare' , figsize = (10,6) , color= '#1fd1a8')
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
plt.title('Mean Age by Sex' , color = 'blue')
plt.ylabel('Average fare' , color = 'blue')
plt.xlabel('Gender' , color= 'blue')
```

```python
# seaborn plots 
# bar plot using seaborn
sns.barplot(data= df , x = 'survived' , y  = 'fare')
```

```python
sns.barplot(data= df , x = 'sex' , y  = 'fare' , hue='class')
```

```python
sns.boxplot(data= df , x = 'sex' , y  = 'age' , hue='class')
```

```python
sns.scatterplot(data= df , x = 'fare' , y  = 'age' , hue='embarked')
```

```python
sns.scatterplot(data= df , y = 'fare' ,  x = 'age' , hue='class')
```

```python
# make scatter plot using plotly
import plotly.express as px

fig = px.scatter(df, x= 'age' , y = 'fare' , color='class')
fig.show()
```

```python
import plotly.graph_objects as go

import pandas as pd

fig = go.Figure(data=[go.Surface(z=df)])

fig.update_layout(title=dict(text='Mt Bruno Elevation'), autosize=False,
                  width=700, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
