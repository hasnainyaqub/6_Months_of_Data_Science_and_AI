# Seaborn and Plotting

```python
# Import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Important things for plotting according to grammer of graphics
1. Data (df)
2. Type of plat
3. x axis
4. y axis
5. Grouping (hue/color) based on categorical variable
6. Size f plot
7. Color palletes of the plot
8. Style of the plot
9. How do you interpret

```python
# Data
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# Type of plot
# Scatter plot
# SCatter theme
sns.set_theme(style='darkgrid')

# Scatter plot size of figure size
plt.figure(figsize=(8, 6))

# change the each scatter point color based on hue
palette = {'male':'blue', 'female':'red'}
sns.scatterplot(data=df , x='age' , y='fare' , hue='sex', size='age'  , palette=palette)

# labels
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatter plot of Age vs Fare")
 
 # Saving a plot 300dpi
# plt.savefig('scatter_plot.png', dpi=300, bbox_inches='tight')
```

Interpertition:
- We have Age of passengers at x-axis and we have fare of passengers at y-axis
- The scatter plot shows the relationship between age and fare of titanic passengers.
- We further grouped passenger by their genders such as males(blue points) and females(red points)
- Two male passengers gave highest fare in titanic and they might be first class passengers.
- The oldest passenger was male & 80 years old and he gave the lowest fare of 7.25 in titanic.

### Box plot 

```python
df['class'].unique()
```

```python
# Type of plot
# box plot
# box theme
sns.set_theme(style='darkgrid')

# box plot size of figure size
plt.figure(figsize=(8, 6))

# change the each scatter point color based on hue
palette = {'Third':'red', 'First':'green', 'Second':'blue'}
sns.boxplot(data=df , x='sex' , y='age' , hue='class' , palette=palette)

# labels
plt.xlabel("sex")
plt.ylabel("age")
plt.title("boxen plot of Age based pn Gender and Class")
 
# Change legend position and name
plt.legend(title='class', loc='upper right')

 # Saving a plot 300dpi
# plt.savefig('scatter_plot.png', dpi=300, bbox_inches='tight')
```

Boxen plot

```python
# Type of plot
# boxen plot
# boxen theme
sns.set_theme(style='darkgrid')

# boxen plot size of figure size
plt.figure(figsize=(8, 6))

# change the each scatter point color based on hue
palette = {'Third':'red', 'First':'green', 'Second':'blue'}
sns.boxenplot(data=df , x='sex' , y='age' , hue='class' , palette=palette)

# labels
plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("boxen plot of Age based pn Gender and Class")
 
# Change legend position and name
plt.legend(title='class', loc='upper right')

# Change the x ticks names
plt.xticks(ticks=[0,1] ,labels=['Male', 'Female'] , rotation=30)


 # Saving a plot 300dpi
# plt.savefig('scatter_plot.png', dpi=300, bbox_inches='tight')
```
