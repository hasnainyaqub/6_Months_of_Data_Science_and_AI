# Data Distribution

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
# silent errors
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import data from seaborn
df = sns.load_dataset('tips')
df.head()
```

```python
# draw a histplot with sns  curve on it 
sns.histplot(data=df, x='total_bill', kde=True)
```

```python
sns.histplot(data=df, x='tip', kde=True)   
```

```python
sns.histplot(data=df, x='size', kde=True)  
```

```python
dft = sns.load_dataset('titanic')
dft.head()
```

```python
# create a fig size big 
# lets create a 15,6 figure size
plt.figure(figsize=(15,6))

sns.histplot(data=dft, x='fare' ,kde=True , bins=300)
```

```python
sns.histplot(data=dft, x='age' ,kde=True, hue='who' , bins=50)
```

```python
sns.boxplot(data=dft , x='age' , hue='who')
```

```python
sns.boxplot(data=dft , x='age' , hue='sex')
```

# Kurtosis & Skewness 

```python
# find kurtoisis of age
stats.kurtosis(dft['fare'], fisher=False)
```

```python
stats.skew(dft['fare'])
```

```python
# plot a histplot of fare
plt.figure(figsize=(10,6))
sns.histplot(data=dft, x='fare', kde=True)
```

```python
# how to test the normal distribution of the data

# H0: data is normal
# H1: data is not normal

# sharpiro wilk test
# D'Agostino's K^2 Test
# Anderson-Darling Test
```

```python
# lets check the sharpiro wilk test
stats.shapiro(dft ['fare'])
```
