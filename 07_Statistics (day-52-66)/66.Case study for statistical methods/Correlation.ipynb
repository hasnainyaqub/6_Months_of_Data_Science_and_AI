# Correlation

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = sns.load_dataset('titanic')
df = df.dropna()
df.head()
```

```python
# create a heatmap of correlation
sns.heatmap(df[['age', 'fare']].corr(), annot=True, cmap='Blues')
```

```python
age_male = df[df['sex'] == 'male']['age']
age_female = df[df['sex'] == 'female']['age']
pearson_corr = age_male.corr(age_female)
spearman_corr = age_male.corr(age_female, method='spearman')
kendall_corr = age_male.corr(age_female, method='kendall')
print(f'Pearson Correlation Coefficient: {pearson_corr}\nSpearman Correlation Coefficient: {spearman_corr}\nKendall Correlation Coefficient: {kendall_corr}')
 
```
