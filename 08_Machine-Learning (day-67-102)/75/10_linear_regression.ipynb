`linear Regression` is a supervised machine learning algorithm where the predicted output is `contionous` and has a constant slope.It's used to predict the value of a dependent variable based on the values of one or more independent variables.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler , MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error , r2_score
```

```python
# load th data tips from sns
df = sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(x='total_bill' , y='tip' , data=df)
```

```python
# split the data into X and y 
X = df[['total_bill' ]]
# ss = MinMaxScaler()
# X= ss.fit_transform(X)
y = df['tip']
```

```python
# split the data into train and test
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2 )
```

```python
# call the model 
model = LinearRegression()
model.fit(X_train , y_train)
```

```python
# take out model intercept and slop, make an equation
print(model.intercept_)
print(model.coef_)
print('y=', model.intercept_ , '+', model.coef_ , 'x')
```

```python
# predict the values
y_pred = model.predict(X_test)
y_pred
```

```python
# model.predict 
model.predict([[30]])
```

```python
# evaluate the model
mse = mean_squared_error(y_test , y_pred)
print(f'MSE = {mse}')
r2 = r2_score(y_test , y_pred)
print(f'R2 = {r2}')
print(f'RMSE = {np.sqrt(mse)}')
```

```python
# plot the model
plt.scatter(X_test , y_test)
plt.plot(X_test , y_pred , color='red')
plt.xlabel('total_bill')
```
