# Stock market data scrapping using yfinance.

```python
# import libraries
import yfinance as yf
import pandas as pd
```

```python
# define ticker symbol 
ticker_symbol = 'GOOGL'
# get data on this ticker
ticker_data = yf.Ticker(ticker_symbol)
# get historical market data
ticker_df = ticker_data.history(period='1d', start='2000-1-1')
# show the data
ticker_df.tail(10)
```

```python
# AAPL ka 1-hour interval data last 60 days ke liye
data = yf.download(tickers='AAPL', period='60d', interval='15m')
print(data.info())
data.head()
```
